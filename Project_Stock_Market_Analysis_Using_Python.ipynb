#Adda new column ‘Trend’ whose values are based on the 'Daily Return' in amazon dataset

```python
amzn=pd.read_csv("/content/AMZN.csv")
amzn.head()
```

```python
amzn_daily_returns=amzn['Adj Close'].pct_change()
amzn_daily_returns.head()
```

```python
amzn_daily_returns.plot(figsize = (15,7))
plt.xlabel("Years")
plt.ylabel("Growth")
plt.title("Daily Returns")
plt.show()
```

```python
amzn["Daily Average Return"]=(amzn["Close"]-amzn["Open"])/amzn["Open"]*100
amzn.head()
```

```python
def trend(x):
  if x <= 0:
    return 'Negative'
  else:
    return 'Positive'
```

```python
amzn["Trend"]=pd.Series(map(trend,amzn["Daily Average Return"]))
```

```python
amzn['Trend']=amzn['Daily Average Return'].apply(lambda x:trend(x))
amzn.head()
```
